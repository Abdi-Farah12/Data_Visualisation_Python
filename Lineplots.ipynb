## Practical activity: Create lineplots



Canopy is a new boutique streaming company that is looking to create an app that provides recommendations based on the last movie you watched. As a part of their process, before they start making suggestions and recommendations to their clients, they want to diagnose the data and gather enough information. 

This analysis uses the **movies.csv** and **ott.xlsx** data sets. Based on the available information, in this activity we will:

- Create a line chart to help identify trends in the data and track/spot interesting relationships among the variables.
- Understand what the data says by visualising it with lineplots.

## 1. Import the libraries

```python
#import necessary packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Import Excel file

```python
#Load the excel data using pd.read_excel
ott = pd.read_excel("ott.xlsx")

print(ott.columns)
```

## 3. Import CSV file

```python
#Load the csv data using pd.read_csv
movies = pd.read_csv("movies.csv")

print(movies.columns)
```

## 4. Validate the DataFrames

```python
# data imported correctly
print(movies.head())
print(movies.shape)
print(movies.dtypes)
```

```python
# data imported correctly
print(ott.head())
print(ott.dtypes)
print(ott.shape)
```

## 5. Combine the two DataFrames

```python
#merge the two DataFrames
mov_ott = pd.merge(movies, ott, how='left', on = "ID")

# view the DataFrame
print(mov_ott.shape)
mov_ott.head()
```

## 6. Create a lineplot

```python
# create a simple lineplot
sns.lineplot(x="Year", y="IMDb", data=mov_ott)
```

```python
# create a simple lineplot
sns.lineplot(x="Year", y="IMDb", data=mov_ott, ci=None)
```

```python
# create lineplots with specification
sns.lineplot(x = "Year", y = "IMDb",
             data=mov_ott[mov_ott["Age"].isin(["16+","18+"])],
             hue ="Age")
```

```python
# create lineplots with specification
sns.lineplot(x = "Year", y = "IMDb",
             data=mov_ott[mov_ott["Age"].isin(["16+","18+"])],
             hue ="Age", ci=None)
```
